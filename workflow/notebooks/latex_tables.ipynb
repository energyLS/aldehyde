```python
import pandas as pd
```

```python

PREFIX = "/mnt/c/Users/scl38887/Documents/git/aldehyde/"
INPUT = PREFIX + "workflow/subworkflows/pypsa-earth-sec/pypsa-earth/resources/powerplants.csv"
OUTPUT = PREFIX + "report/sections/tables/powerplants.tex"
```

```python
powerplants = pd.read_csv(INPUT, index_col=0)
```

```python
powerplants.head()
```

```python
keep = ["Name", "Capacity", "Fueltype", "DateIn", "DateOut"]
powerplants = powerplants[keep]
```

```python
test = powerplants.reset_index(drop=True)
```

```python
test.head()
```

```python
test["Capacity"] = test["Capacity"].round(1)
```

```python
test["DateIn"] = test["DateIn"].astype(int)
test["DateOut"] = test["DateOut"].astype(int)
```

```python
# Sort by capacity
test = test.sort_values(by="Capacity", ascending=False)
```

```python
# Rename columns
test = test.rename(columns={"Capacity": "Capacity in MW", "Fueltype": "Fuel type", "DateIn": "Commission", "DateOut": "Decommission"})
```

```python
test.to_latex(
    OUTPUT,
    longtable=False,
    column_format="p{1cm}p{6.5cm}rrp{2cm}p{3.4cm}",
    na_rep="--",
    escape=False,
    label="tab:powerplants",
    caption="Conventional power plants in Morocco",
)
```
